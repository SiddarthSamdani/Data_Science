```python
import pandas as pd
import numpy as np
import seaborn as sns

data = pd.read_csv("movies.csv")
data.head(2)
```

```python
sns.countplot(x="country", hue="language", data=data)
sns.set(rc={'figure.figsize':(80,9)})
```

```python
sns.relplot(x="budget", y="gross", data=data)
```

```python
sns.scatterplot(x="budget", y="gross",hue="country", data=data)
sns.set(rc={'figure.figsize':(20,9)})
```

```python
sns.lineplot(x="budget", y="gross",hue="country", data=data)
sns.set(rc={'figure.figsize':(20,9)})
```

```python
sns.catplot(x="budget", y="gross",hue="country", data=data)
sns.set(rc={'figure.figsize':(20,9)})
```

```python
sns.stripplot(x="budget", y="gross",hue="country", data=data)
sns.set(rc={'figure.figsize':(20,9)})
```

```python
sns.swarmplot(x="budget", y="gross",hue="country", data=data)
sns.set(rc={'figure.figsize':(20,9)})
```

```python
sns.boxplot(x="budget", y="gross",hue="country", data=data)
sns.set(rc={'figure.figsize':(20,9)})
```

```python
sns.violinplot(x="budget", y="gross",hue="country", data=data)
sns.set(rc={'figure.figsize':(20,9)})
```

```python
sns.boxenplot(x="budget", y="gross",hue="country", data=data)
sns.set(rc={'figure.figsize':(20,9)})
```

```python
sns.pointplot(x="budget", y="gross",hue="language", data=data)
sns.set(rc={'figure.figsize':(20,9)})
```

```python
sns.barplot(x="budget", y="gross",hue="country", data=data)
sns.set(rc={'figure.figsize':(20,9)})
```
